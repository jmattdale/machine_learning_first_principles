```python
import numpy as np
import matplotlib.pyplot as plt
```

### Binary CrossEntropy Cost Function

$$J(\theta) = -\frac{1}{m} \sum_{i=1}^m y^{(i)}\log (h(z(\theta)^{(i)})) + (1-y^{(i)})\log (1-h(z(\theta)^{(i)}))\tag{5} $$

#### Vectorized Form
$$h = g(X\theta) $$
$$J(\theta) = \frac{1}{m} \cdot \Bigl(-y^T \log(h) - (1-y)^T\log(1-h)\Bigr)$$


Crossentropy loss aims to weight the cost of mismatches between a probability scores [0-1] and ground truth values [0, 1].
 - In the case of a mismatch ie **score: 0.999, ground truth: 0.0**
   - Loss approaches infinity
 - In the case of a match ie **score: 0.999, ground truth: 1.0**
   - Loss approaches 0.0

```python
def binary_crossentropy_loss(y_pred, y_true):
    y_pred, y_true = np.array(y_pred), np.array(y_true)
    assert y_pred.shape[0] == y_true.shape[0]
    
    positive_loss = y_true * -np.log(y_pred) # -log(probability)
    
    negative_loss = (1 - y_true) * -np.log(1 - y_pred) #
    
    return positive_loss + negative_loss
```

```python
losses = [
    binary_crossentropy_loss([i], [1.0]) 
    for i in np.linspace(0.0, 1.0)[::-1]
]

plt.plot(np.linspace(1, 0), losses)
plt.ylabel('loss')
plt.xlabel('probability')
plt.title('Log loss for positive ground truth value');
```

```python
losses = [
    binary_crossentropy_loss([i], [0.0]) 
    for i in np.linspace(0.0, 1.0)[::-1]
]

plt.plot(np.linspace(1, 0), losses)
plt.ylabel('loss')
plt.xlabel('probability')
plt.title('Log loss for negative ground truth value');
```
